```python
import matplotlib.pyplot as plt
import math

def compute_lambda(mean_time):
    return 1 / mean_time

def exponential_pdf(x, lambda_rate):
    """Compute the exponential probability density function."""
    return lambda_rate * math.exp(-lambda_rate * x)

def generate_distribution(mean_time, step=0.1, max_time=10):
    """Generate x and y values for exponential distribution."""
    lambda_rate = compute_lambda(mean_time)
    x_values = [x * step for x in range(int(max_time / step))]
    y_values = [exponential_pdf(x, lambda_rate) for x in x_values]
    return x_values, y_values

def plot_distribution(x, y, title="Exponential Distribution", xlabel="Time", ylabel="Probability Density"):
    """Plot the distribution."""
    plt.figure(figsize=(10, 6))
    plt.plot(x, y, color='green', linewidth=2)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()
```

```python
mean_time = 5  # average time between web requests in seconds
x_vals, y_vals = generate_distribution(mean_time)
plot_distribution(x_vals, y_vals, title="Exponential Distribution - Time Between Web Requests",
                  xlabel="Time Between Requests (seconds)",
                  ylabel="Probability Density")
```
